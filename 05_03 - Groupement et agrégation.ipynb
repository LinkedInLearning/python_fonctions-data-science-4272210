<h3 style="text-align: left; background-color: #e0e0e0; font-family:Times New Roman; color: black; padding: 14px; line-height: 1; border-radius:10px">Groupement et agrégation</h3>

```python
import polars as pl
```

```python
df = pl.read_csv("Dataset_Employes.csv", sep=";")
df
```

```python
grouped_df = df.groupby('Contrat').agg(pl.all().mean())
grouped_df
```

```python
grouped_df = df.groupby('Contrat').agg(pl.col('%variable_moyen').mean())
grouped_df
```
