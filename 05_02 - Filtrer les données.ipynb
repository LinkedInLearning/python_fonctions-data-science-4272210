```python
df.select('Age')
```

```python
df.filter(pl.col('Age') < 35)
```

```python
df.filter(pl.col('Contrat') == "CDD")
```

```python
df.filter((pl.col('Contrat') == "CDD") & (pl.col('Age') < 35))
```
