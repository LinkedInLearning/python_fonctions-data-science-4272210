```python
import pandas as pd
```

```python
df = pd.read_excel("Dataset_Employes.xlsx")
df
```

```python
# Modifier un élément du tableau
df.at[0, 'Contrat'] = "CDD"
df
```

```python
# Créer une nouvelle colonne
df['Ratio_satisfaction']=df['Niveau de satisfaction']/100
df
```

```python
# Renommer une colonne
df.rename(columns={'Ancienneté_an': 'Annnée_Ancienneté'}, inplace=True)
df
```

```python
# Supprimer une colonne
df = df.drop(columns=['Niveau de satisfaction', 'id_salarié'])
df
```

```python
# Modifier des valeurs
df['Type Contrat'] = df['Type Contrat'].replace('Non Cadre', 'Technicien')
df
```
