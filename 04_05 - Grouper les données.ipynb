<h3 style="text-align: left; background-color: #e0e0e0; font-family:Times New Roman; color: black; padding: 14px; line-height: 1; border-radius:10px">Grouper les données</h3>

```python
import pandas as pd
```

```python
df = pd.read_excel("Dataset_Employes.xlsx")
df
```

```python
# Grouper les données par 'Contrat' et calcul de la moyenne des salaires
grouped = df.groupby('Contrat')['Salaire base mensuel'].mean()
print(grouped)
```

```python
# Grouper les données par 'Contrat' ET pas 'Type de Contrat' et calcul de la moyenne des salaires
grouped = df.groupby(['Contrat', 'Type Contrat'])['Salaire base mensuel'].mean()
print(grouped)
```

```python
# Appliquer plusieurs fonctions d'agrégation (médiane et moyenne)
agg_result = df.groupby(['Contrat', 'Type Contrat'])['Salaire base mensuel'].agg(['median', 'mean'])
print(agg_result)
```
