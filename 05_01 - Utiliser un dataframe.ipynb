Polars est une bibliothèque de manipulation de données hautement performante en Python, similaire à Pandas, mais optimisée pour la vitesse et l'utilisation de la mémoire, en particulier pour les grandes quantités de données.  

Importer un fichier Excel dans un DataFrame Polars nécessite quelques étapes, car Polars ne prend pas directement en charge les fichiers Excel, contrairement aux fichiers CSV. 

```python
import pandas as pd
```

```python
#!pip install polars
```

```python
import polars as pl
```

```python
df_pandas = pd.read_excel("Dataset_Employes.xlsx")

# Convertir le DataFrame Pandas en DataFrame Polars
df_polars = pl.from_pandas(df_pandas)
df_polars
```
